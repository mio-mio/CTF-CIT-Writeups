## Challenge Overview
- **Category:** Forensics
- **Challenge Name:** True CTF Love

## Description
- **Challenge Text:**  
  I got this strange email from another CTF participant not too long ago. I am just not sure what they mean by this...  
  Do you love CTFs as much as they do?  
- **Data:** The_Flag_Well_Capture_Together.eml

スクリーンショット 2025-04-29 14.42.04.png




## Initial Analysis
- The file format was `.eml`, a standard email message format.
- Opened the file and noted it was a heartfelt poem related to CTFs.
- The subject line was encoded in Base64:  
  `=?UTF-8?Q?The_Flag_We=E2=80=99ll_Capture_Together?=`  
  which decodes to: `The Flag We’ll Capture Together`.
- The email body contained no immediate flag-like patterns.

## Attempts and Progress
- Used `strings` and `grep` to search for common flag formats like `CIT{...}` but no matches found.
- Looked for hidden attachments or encoded blobs within the body or headers、but no clue.
- Observed that the email seemed too poetic and clean to contain traditional hidden data.
- Eventually re-examined the filename more carefully.



## Solution

スクリーンショット 2025-04-29 14.54.37.png

- After learning several sources, I checked the file with `less The_Flag_Well_Capture_Together.eml` and found a second `b=` field in the DKIM-Signature.
- Normally there is only one `b=`, which contains a signature that can be verified using the sender's pulbic key. However, there was a second `b=` field was present.
- The second `b=` field contained the string: `
V293LCB3aGF0IGEgYmVhdXRpZnVsIGxpdHRsZSBwb2VtLiBJIGFsbW9zdCBzaGVkIGEgdGVhciByZWFkaW5nIHRoYXQuIEhvcGVmdWxseSB5b3UgbGVhcm5lZCBtb3JlIGFib3V0IGVtYWlsIGhlYWRlcnMuIEJ1dCBzZXJpb3VzbHksIGl0IGdldHMgbWUgd29uZGVyaW5nLi4uIGRvIHlvdSBsb3ZlIENURnMgYXMgbXVjaCBhcyB0aGV5IGRvPwoKQ0lUe2lfbDB2M19jdGYkX3QwMH0`
- This string was Base64-encoded. After decoding, the flag was revealed.
