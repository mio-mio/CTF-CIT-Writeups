## Challenge Overview
**Title:** Blank Image  
**Category:** Steganography

## Description
> I was gonna make a really cool challenge but then I literally forgot about it so all I have is this blank image. Good luck!

Provided data: `image.png`



スクリーンショット 2025-04-27 21.15.02.png




## Initial Analysis
- Opened `image.png`, which appeared completely blank (no visible content).
- The challenge description implied that the flag could be hidden inside the image at the byte level.

## Attempts and Progress
- First tried basic checks:
  - `file image.png` confirmed it was a valid PNG file.
  - `strings image.png` and `xxd image.png` did not reveal any readable flag.
  - `exiftool image.png` found no helpful metadata.

- Suspected steganography techniques.
  - `binwalk image.png`, `foremost -i image.png`, `pngcheck -v image.png` did not reveal any clue.

## Solution
- Used `zsteg`, a specialized tool for PNG/BMP steganography:
  `zsteg image.png` directly revealed the hidden flag in ASCII text.


スクリーンショット 2025-04-27 21.18.35.png


## Reflection
-     Learned that even a blank image can conceal steganographic data inside its least significant bits (LSBs).

- zsteg proved to be extremely effective for this kind of quick byte-level analysis.

- In future challenges, will prioritize using zsteg earlier when facing suspiciously "empty" images.
