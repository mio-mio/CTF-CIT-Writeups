## Challenge Overview
**Title:** Queen's Gambit  
**Category:** Steganography

## Description
> Like the actual Queen, not the band.

Provided data: `Queen.png`

---

## Initial Analysis
- Opened `Queen.png`, confirmed it was a valid PNG file.

## Attempts and Progress
- Performed a systematic steganographic analysis using the following methods:
- Verified file type with `file Queen.png`.
- Searched for hidden text using `strings` and `grep`.
- Scanned for embedded data with `binwalk`.
- Attempted file recovery with `foremost`.
- Checked file integrity with `pngcheck -v`.
 - No clues are found so far.

- Ran `stegsnow` to detect hidden text using spaces and tabs:
  - Found "HA" as output, likely a decoy.
  - Decompression with `-C` option raised a warning (residual bits), but no usable data.


## Solution


スクリーンショット 2025-04-29 13.28.26.png

- Used `zsteg Queen.png` to find suspicious text: "a8 a7 a6 a5 b8 c7 b6 d5 e4 f5 g4 h5 c1 c2 c3 d2 e1 e2 e3".
- According to other sources, this text represents positions on a chessboard.
- When marking each position one by one, the letters "P", "w", and "n" appear, forming the flag.
