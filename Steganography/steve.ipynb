## Initial Analysis

  - It looked like an image of the scene from Minecraft movie.
- The challenge title and description suggested a possible Minecraft reference, but the main hint seemed steganographic rather than game-based.

## Attempts and Progress
- Used `file ChickenJockey.png` to confirm the file type:
  - Confirmed as a standard PNG image.

- Performed basic checks:
  - Used `strings ChickenJockey.png` to search for readable text.
  - Ran `binwalk ChickenJockey.png` to find embedded files or code.
  - Used `pngcheck -v ChickenJockey.png` to check for suspicious errors.
  - Used `exiftool ChickenJockey.png` to inspect metadata.
- None of these methods revealed a clear flag.

## Solution
- When I used `zsteg ChickenJockey.png`, I found a suspicious string: `VEhJU19pc19hX2NyYWZ0aW5nX3RhYmxl`.
- I tried decoding it with **Base64**, and it turned out to be the flag.

## Reflection
- Learned that when encountering suspicious-looking encoded strings, it is important to quickly try basic decoding methods like Base64.
