## Challenge Overview
- **Category:** Crypto
- **Challenge Name:** Nerd Emoji

## Description
- **Challenge Text:**  
  Find the flag.  
  **Data:** nerdemoji.txt

## Initial Analysis
- The file `nerdemoji.txt` consisted of long sequences of seemingly nonsensical words. This suggested the use of some classical cipher that required decoding.

## Attempts and Progress
- Upon the inspection, a flag-like string was found at the end of the file:  
  `EOZ{BLtysIZkEiJQwRULrV}`
- Various common ciphers such as ROT, Vigenère, Caesar, and Autokey were tested but yielded no valid results.

## Solution
- By assuming that EOZ corresponds to CIT, and that recurring words like Zit = The, and q = a, we inferred that a simple substitution cipher was used.

- Using a substitution cipher tool with these clues allowed successful decryption of the flag.
