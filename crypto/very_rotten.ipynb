## Challenge Overview
- **Category:** Crypto
- **Challenge Name:** Very Rotten

## Description
- **Challenge Text:**  
  `TFJDe0gzeV9ldmlfdzMzX0FYQ0M2V19Wa1hidldINDYxTXR1YlgyOXZnYn0=`  

## Initial Analysis
- The given string is clearly Base64-encoded.
- The challenge name "Very Rotten" hints at use of the **ROT cipher**, possibly ROT13 cipher.

## Attempts and Progress
- Decoding the initial string once produced a flag-like format:  
  `LRC{H3y_evi_w33_AXCC6W_VkXbvWH461MtubX29vgb}`  
  suggesting further decoding was likely be needed.
- Attempted ROT13, ROT47, and variations of Caesar shift on the decoded result.
- Tested full ROT transformations to restore the `CIT{...}` format, but no readable words were found.

## Solution
- While reviewing other sources, I discovered that a different rotation amount was required to decode the flag.
- Applying a Caesar cipher with a rotation of 17 to uppercase letters and digits, and a rotation of 22 to lowercase letters, successfully revealed the correct flag.
