## Description
- **Challenge Text:**  
`PFFUQYTUONPSK5LAMNDXGJ35ER4CM5C7ORETY3A=`

## Initial Analysis
- The challenge text appears to be encoded in BASE64 or BASE32.
- XOR decryption is likely involved at some stage.

## Attempts and Progress
- Ran `echo 'PFFUQYTUONPSK5LAMNDXGJ35ER4CM5C7ORETY3A=' | base32 -d` and tried 1-byte XOR brute force on the decoded result, but no flag can be found.
- Tried cipher identifier and experimented with `Base91 Encodiing`, `ASCII Shift Cipher` or `ROT-47 Cipher`.
- After applying `ROT-47` decoding, ran XOR brute force again but still found nothing meaningful.

## Solution

- While reviewing other sources, found that the correct decoding chain was to first apply ROT-47 to the Base32-decoded string, then decode the result using Base64, and finally try XOR decryption using the key `CIT{`.
- Tried some XOR and found suspicious output start from `duck`
- Applied `duck` as a key and tried XOR lead to the flag.

## Reflection
- Even short, encoded strings can hide multiple layers of transformation.

- Learned the importance of decoding steps in the correct order—just using the right tools isn't enough if the logic is wrong.
