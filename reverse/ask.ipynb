## Challenge Overview
- **Category:** Reverse Engineering
- **Challenge Name:** Ask Nicely

## Description
- **Challenge Text:**  
  I made this program, you just have to ask really nicely for the flag!  
  **Data:** asknicely


スクリーンショット 2025-04-30 12.40.58.png



## Initial Analysis
- The file was a statically linked ELF 64-bit binary named `asknicely`.
- Executing the binary prompted the user with:  
  `How badly do you want the flag?`
- User input was required, but typical polite phrases like "please", "pretty please", etc., did not yield the flag.
- prompt keep asking with:  
  `Ask nicely...`
- User input was required again, but typical polite phrases did not yield the flag again.  The prompt ends with:  
  `that's not quite what I'm lookng for.`
- The name and prompt suggested the challenge was about input matching or string comparison.

## Attempts and Progress
- Tried various polite phrases as input, with different casing and punctuation.
- Ran `gdb` to examine the binary more closely, but the error happened like below and couldn't find any clue.  
  `warning: BFD:/mnt/asknicely: invalid string offset 4194304 >= 297077 for section .strtab'`

## Solution

スクリーンショット 2025-04-30 13.16.57.png


- While reviewing additional information, I learned that GUI-based binary analysis tools can be helpful.
- Opened the file with `BinaryNinja` and searched for the string `How badly do you want the flag?`, and I found:  
  `pretty pretty pretty pretty pretty please with sprinkles and a cherry on top`
- Entering this phrase revealed the flag.
